```python
#importing  the require libraries
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

 # Load Dataset 

```python
 
boston =load_boston()
#Description of the dataset.
print(boston.DESCR)
```

# step1 : Data Preprocessing

```python
# put the data into pandas dataframe
features =pd.DataFrame(boston.data,columns=boston.feature_names)
features
```

```python
features['AGE']
```

```python
features['DIS']
```

```python
target =pd.DataFrame(boston.target,columns=['target'])
target
```

```python
max(target['target'])
```

```python
min(target['target'])
```

```python
# Concentates fatures and target into singal target
#axis=1 makes it concentates
df =pd.concat([features,target], axis=1)
df
```

```python

df.describe().round (decimals = 2)
```

#  step2 : Data Visualization

```python
# step2 : Data Visualization
# correlation between every attribute and target
corr = df.corr('pearson')
# to Take absoulte value for correlation
corrs =[abs(corr[attr]['target']) for attr in list(features)]
# make list of the pairs
l =list(zip(corrs,list(features)))

# sort 
l.sort(key=lambda x:x[0], reverse=True)

#unzip
corrs, labels  = list(zip((*l)))

# plot correlation
index = np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index, corrs, width=0.5)
plt.xlabel('Attributes')
plt.ylabel('correlation with target value')
plt.xticks(index, labels)
plt.show()
```

# Normalization of BH Data

```python
X=df['LSTAT'].values
Y=df['target'].values
Y[:5]
```

```python
x_scaler=MinMaxScaler()
x=x_scaler.fit_transform(X.reshape(-1 ,1))
x=x[:,-1]
y_scaler=MinMaxScaler()
y=y_scaler.fit_transform(Y.reshape(-1 ,1))
y=y[:,-1]
```

```python
print(y[:5])
```

# Step:3 Splitting the Data:

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
xtrain = xtrain.reshape(-1,1)
ytrain = ytrain.reshape(-1,1)
xtest = xtest.reshape(-1,1)
ytest = ytest.reshape(-1,1)
```

# Step 4: Training Model

```python
# define regression object
lm=LinearRegression()
```

```python
# fitting the model
lm.fit(xtrain,ytrain)
```

```python
# parameter of the feature
lm.coef_
```

# Step :5 Making Prediction 

```python
#prediction
prediction=lm.predict(xtest)
```

```python
#plotting 
plt.scatter(ytest, prediction)
plt.xlabel(' Y test')
plt.ylabel('predictions')
```

```python
from sklearn import metrics
print('MSE: ',metrics.mean_squared_error(ytest,prediction))
```

```python
p=pd.DataFrame(list(zip(xtest,ytest,prediction)),columns = ['x','target_y','predictions'])
p
```

# Ploting the predicted values against the target values

```python
plt.scatter(xtest,ytest,color='b')
plt.plot(xtest,prediction,color='r')
plt.show()
```

# Step 6: Reverting Normalization to Final obtain prediction of the houses price

```python
prediction=prediction.reshape(-1,1)
xtest = xtest.reshape(-1,1)
ytest = ytest.reshape(-1,1)

xtest_scaled =x_scaler.inverse_transform(xtest)
ytest_scaled =y_scaler.inverse_transform(ytest)
prediction_scaled =y_scaler.inverse_transform(prediction)

xtest_scaled = xtest_scaled[:,-1]
ytest_scaled = ytest_scaled[:,-1]
prediction_scaled = prediction_scaled[:,-1]

p=pd.DataFrame(list(zip(xtest_scaled,ytest_scaled,prediction_scaled)),columns = ['x','target_y','predictions'])
p =p.round(decimals= 2)
p
```
